## Task 2: Quantify each signal

For each trace, compute:
- mean
- standard deviation
- max absolute value

```python
signals = [c for c in df.columns if c != "t_s"]
summary = []
for c in signals:
    s = df[c]
    summary.append({
        "signal": c,
        "mean_m": s.mean(),
        "std_m": s.std(),
        "max_abs_m": s.abs().max()
    })

summary_df = pd.DataFrame(summary)
summary_df
```
